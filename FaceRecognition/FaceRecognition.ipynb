```python
import cv2
import matplotlib.pyplot as plt
```

```python
model=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
img=cv2.imread("Front_face_photo.jpg")
```

```python
plt.imshow(img)
```

```python
newimg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(newimg)
```

```python
faces=model.detectMultiScale(newimg,1.3,1)
```

```python
faces
```

```python
for i in faces:
    x,y,w,h=i
    cv2.rectangle(newimg,(x,y),(x+w,y+h),(0,255,0),2)

plt.imshow(newimg)
```

```python
cam=cv2.VideoCapture(0)
while True:
    success,img=cam.read()

    faces=model.detectMultiScale(img,1.2,5)

    for f in faces:
        x,y,w,h=f
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)

        cv2.imshow("Image Window",img)
        key=cv2.waitKey(1)

    if(key==ord('q')):
        break

cam.release()
cv2.destroyAllWindows()

```
